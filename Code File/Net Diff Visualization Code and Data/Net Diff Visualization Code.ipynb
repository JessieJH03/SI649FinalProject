```python
import pandas as pd
import altair as alt


df_23 = pd.read_excel('2022-2023 Celtics Season.xlsx') 
df_24 = pd.read_excel('2023-2024 Celtics Season.xlsx') 

# Merge the data and Only keep G、Plus/Miuns_23 和 Plus/Miuns_24

df_combined = pd.merge(df_23[['G', 'Plus/Minus_23']], df_24[['G', 'Plus/Minus_24']], on='G')

# convert to long format
df_long = pd.melt(
    df_combined,
    id_vars=['G'],  
    value_vars=['Plus/Minus_23', 'Plus/Minus_24'],  
    var_name='Season',
    value_name='Plus/Minus'
)

# Clean Season 
df_long['Season'] = df_long['Season'].str.replace('Plus/Minus_', '')


zoom = alt.selection_interval(bind='scales', encodings=['x'])
nearest = alt.selection_point(on='mousemove', fields=['G'], nearest=True)

line_chart = alt.Chart(df_long).mark_line(size=2).encode(
    x=alt.X('G:Q', title='Game Number'),
    y=alt.Y('Plus/Minus:Q', title='Plus/Minus'),
    color=alt.Color('Season:N', title='Season')
).add_params(zoom)

# Vertical Reference Line
vline = alt.Chart(df_long).mark_rule(color='lightgray', size=1).encode(
    x='G:Q'
).transform_filter(nearest)


points = alt.Chart(df_long).mark_circle(size=80, filled=True).encode(
    x='G:Q',
    y='Plus/Minus:Q',
    color=alt.Color('Season:N'),
    opacity=alt.condition(nearest, alt.value(1), alt.value(0))
).add_params(nearest)


text = points.mark_text(align='left', dx=7, fontSize=10).encode(
    text=alt.Text('Plus/Minus:Q', format='.1f')
)

# Merge the layer
chart = alt.layer(line_chart, vline, points, text).properties(
    width=800,
    height=400,
    title="Plus/Minus difference by game of Celtics in 22-23 and 23-24 Season"
).configure(
    background='white'
).configure_axis(
    labelFontSize=12,
    titleFontSize=14,
    domainColor='black',
    tickColor='lightgray',
    labelColor='black',
    gridColor='lightgray',
    titleColor='black'
).configure_title(
    fontSize=16,
    color='black'
).configure_legend(
    labelColor='black',
    titleColor='black'
)

chart
```
